```python
# 1). Read Salaries.csv into a pandas dataframe

import pandas as pd
import numpy as np

df = pd.read_csv('Salaries.csv')
df
```

```python
# 2). Create a column named 'EmployeeFirstName' and extract the first name from the EmployeeName Column

df['EmployeeFirstName'] = df['EmployeeName'].str.split().str[0]
df
```

```python
# 3). Create a column named 'EmployeeLastName' and extract the last name from the EmployeeName Column

df['EmployeeLastName'] = df['EmployeeName'].str.split().str[-1]
df
```

```python
# 4). Convert all letters in the JobTitle column to all upper case

df['JobTitle'] = df['JobTitle'].str.upper()
df
```

```python
# 5). Find the average TotalPay of a transit operator

df_transit_officer = df[df['JobTitle'].str.contains('TRANSIT')]
df_transit_officer['TotalPay'].mean()
```

```python
# 6). Find the top 10 employees with the highest Total pay

df.sort_values(['TotalPay'], ascending=False).head(10)
```

```python
# 7). Find all employees where there OtherPay exceeds $100,000

df[df.OtherPay > 100000]
```

```python
# 8). Find the mean TotalPay where the JobTitle contains the word "Police" 

df_police = df[df['JobTitle'].str.contains('POLICE')]
df_police['TotalPay'].mean()
```

```python
# 9). Find the  person with the Max TotalPay where the JobTitle contains the word "Police"

df_police.sort_values(['TotalPay'], ascending = False).head(1)
```

```python
# 9b) Verifying max TotalPay using max():

df_police['TotalPay'].max()
```

```python
# 10) Find the highest paid nurse.

df_nurses = df[df['JobTitle'].str.contains('NURSE')]
df_nurses.sort_values(['TotalPay'], ascending = False).head(1)
```

```python
# 10b) Verifying max TotalPay using max():

df_nurses['TotalPay'].max()
```
