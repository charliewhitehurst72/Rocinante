```python
# Make a two-player Rock-Paper-Scissors game. 
# (Hint: Ask for player plays (using input), compare them, print out a message of congratulations to the winner, 
# and ask if the players want to start a new game)
# 
# Remember the rules:
#
# Rock beats scissors
# Scissors beats paper
# Paper beats rock

print('Please choose one of the following options: \n Rock\n Paper\n Scissors\n')
active = True
while active:
    play_options = {'ROCK': 1, 'SCISSORS': 2, 'PAPER': 3}
    player_1 = input('Player 1, what is your choice? ').upper()
    player_2 = input('\nPlayer 2, what is your choice? ').upper()
    player_1_choice = play_options.get(player_1)
    player_2_choice = play_options.get(player_2)
    win_value = player_1_choice - player_2_choice
    
    if win_value in [-1, 2]:
        if input("\nPlayer 1 is the winner! \nWould you like to play again? \nEnter 'y' for yes or anything else for no: ").upper() == 'Y':
            continue
        else:
            print('\nThanks for playing!')
            break
        
    elif win_value in [-2, 1]:
        if input("\nPlayer 2 is the winner! \nWould you like to play again? \nEnter 'y' for yes or anything else for no: ").upper() == 'Y':
            continue
        else:
            print('\nThanks for playing!')
            break
        
    else:
        print('\nThe result is a tie. Please choose again. ')
```
