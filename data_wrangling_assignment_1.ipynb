```python
# df['email'] = np.where(~df['email'].str.contains("@", na=False), "N/A", df['email'])
# Peter's code for the email questions ^

# 1). Import Pandas and read the customers.csv into a pandas dataframe

import pandas as pd
import numpy as np

customers = pd.read_csv('customers.csv')
customers
```

```python
# 2). Check to see if the "id" column is a unique identifier.

customers['id'].is_unique
```

```python
# 3). Create a unique identifier column titled "customerID" and set this column as the index

customers['customerID'] = customers.index
customers = customers.set_index('customerID')
customers
```

```python
# 4) Drop the "id" column

customers.drop(['id'], axis = 1)
```

```python
# 5). Parse the column "full_name" and create the columns first_name and last_name.

full_name = customers['full_name'].str.split(' ', n = 1, expand = True)
customers['first_name'] = full_name[0]
customers['last_name'] = full_name[1]
customers
```

```python
# 6). Drop the full_name column

customers.drop(['full_name'], axis = 1)
```

```python
# 7.) Display all emails that are missing a '@'

customers[customers["email"].str.contains('@')==False]
```

```python
# 8.) Replace all emails that are missing a "@" with "N/A".

customers['email'] = np.where(~customers['email'].str.contains("@", na=False), "N/A", customers['email'])
customers
```

```python
# 9.) Parse the column "city" to create the columns "city", "state", and "zip"

customers[['City', 'State']] = customers['city'].str.split(',', 1, expand = True)
customers['Zip'] = customers['State'].str[-5:]
customers['State'] = customers['State'].astype(str).str[:-6]
customers
```

```python
# 10.) Drop the original "city" column

customers.drop(['city'], axis = 1)
```

```python
# 11.) Rename the column "shirt size" to "shirt_size"

customers.rename({'shirt size': 'shirt_size'}, axis = 1, inplace = True)
customers
```
