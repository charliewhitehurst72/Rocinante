```python
# 1). Import Pandas, Numpy and read cars.csv into a pandas dataframe

import pandas as pd
import numpy as np
df = pd.read_csv('cars.csv')
df
```

```python
# 2). Return a count of how many times each car manufacturer appears in the data. 
# Order your results in descending order. 

df_cars_by_manufacturer = df.groupby(['car_make'])['car_vin'].count().sort_values(ascending = False)
df_cars_by_manufacturer
```

```python
# 3). Return a count of how many cars are in each state. Order your results in descending order.

df_cars_by_state = df.groupby(['state'])['car_vin'].count().sort_values(ascending = False)
df_cars_by_state
```

```python
# 4). Filter the dataframe to show only cars made by Lexus where the model is ES

df2 = df[(df['car_make'] == 'Lexus') & (df['car_model'] == 'ES')]
df2
```

```python
# 5). Filter the dataframe to show only fords made after the year 2000 in the state of California

df3 = df[(df['car_make'] == 'Ford') & (df['car_model_year'] > 2000)]
df3
```
